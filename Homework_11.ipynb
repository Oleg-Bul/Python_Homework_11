f(x) = sin(x)^2 - cos(x)^2

1.Определить корни

2.Найти интервалы, на которых функция возрастает

3.Найти интервалы, на которых функция убывает

4.Построить график

5.Вычислить вершину

6.Определить промежутки, на котором f > 0

7.Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy import pprint, S
from sympy.plotting import plot
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
```

```python
# 4.Построить график
plot(f)
```

```python
# 1 Определить корни
ans = solve(f, x)
print(ans)
for i in ans:
    print(i.evalf())
```

```python
# 6. Определить промежутки, на котором f > 0
pprint(solveset(f > 0, x, domain=S.Reals), use_unicode=True)
solve(f > 0, x)
```

```python
# 7. Определить промежутки, на котором f < 0
pprint(solveset(f < 0, x, domain=S.Reals), use_unicode=True)
solve(f < 0, x)
```

```python
# 5  Вычислить вершину
diff_f = diff(f, x)
print(diff_f)
solve(diff_f, x)
```

```python
# 2 Найти интервалы, на которых функция возрастает
pprint(solveset(diff_f < 0, x, domain=S.Reals), use_unicode=True)
```

```python
# 3 Найти интервалы, на которых функция убывает
pprint(solveset(diff_f > 0, x, domain=S.Reals), use_unicode=True)
```
